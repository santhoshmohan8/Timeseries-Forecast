```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

df = pd.read_csv("ice_cream.csv")
df.head()

df.isna().sum()
df.shape

df.rename(columns={'IPN31152N':'qty', 'DATE':'date'}, inplace=True)
df['date'] = pd.to_datetime(df.date)
df.set_index('date', inplace=True)
df.head()
#test
# filtering for subset of data
start_date = pd.to_datetime('2010-01-01')
df = df.loc[start_date:,]

plt.figure(figsize=(10,6))
plt.plot(df)
plt.title("Ice cream production by year")
plt.ylabel("Quantity")
for year in range(start_date.year,df.index.max().year):
  plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', alpha=0.1)
```

Based on the diminishing behavior of ACF, we are likely dealing with AR type of model

```python
plot_acf(df.qty, lags = 100)
```

```python
PACF = plot_pacf(df.qty, lags=50)
```
