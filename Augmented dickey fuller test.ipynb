```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

```python
def generate_ar_data(lags, coeff, length):
  coefs = np.array(coeff)
  series = [np.random.normal() for _ in range(lags)]
  for _ in range(length):
    prev_value = series[-lags:][::-1]
    new_value = np.sum(np.array(prev_value)*coefs) + np.random.normal()
    series.append(new_value)
  return np.array(series)
```

```python
ar1 = generate_ar_data(1,[0.5],100)
plt.figure(figsize=(10,6))
plt.plot(ar1)
plt.title("AR1 process - Stationary data")
```

```python
print("Global mean : ",ar1.mean(),"\nLocal mean 1 : ",ar1[0:19].mean(),"\nLocal mean 2 : ",ar1[20:40].mean(),"\nLocal mean 3 : ",ar1[60:80].mean())

print(['Series is stationary' if adfuller(ar1)[1] < 0.05 else 'Series is not stationary'])
```

```python
ar1 = generate_ar_data(1,[1],100)
plt.figure(figsize=(10,6))
plt.plot(ar1)
plt.title("AR1 process - Non Stationary data")
```

```python
print("Global mean : ",ar1.mean(),"\nLocal mean 1 : ",ar1[0:19].mean(),"\nLocal mean 2 : ",ar1[20:40].mean(),"\nLocal mean 3 : ",ar1[60:80].mean())
print(['Series is stationary' if adfuller(ar1)[1] < 0.05 else 'Series is not stationary'])
```
